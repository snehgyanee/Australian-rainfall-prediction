```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['year'] = df.Date.dt.year
df['month'] = df.Date.dt.month
df['day'] = df.Date.dt.day
```

```python
df.head()
```

```python
df = df.drop('Date',axis=1)
```

```python
df.info()
```

```python
obj = (df.dtypes == "object")
object_cols = list(obj[obj].index)
```

```python
object_cols
```

```python
for i in object_cols:
    print(i,df[i].isnull().sum())
```

```python
for i in object_cols:
    df[i].fillna(df[i].mode()[0], inplace=True)
```

```python
numerical = (df.dtypes == 'float64')
num_cols = list(numerical[numerical].index)
```

```python
num_cols
```

```python
for i in num_cols:
    print(i,df[i].isnull().sum())
```

```python
for i in num_cols:
    df[i].fillna(df[i].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in object_cols:
    df[i] = le.fit_transform(df[i])
```

```python
df.info()
```

```python
df2 = df[['RainToday','RainTomorrow','year','month','day']]
```

```python
df3 = df[['Location', 'MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
       'WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
scaled_data = sc.fit_transform(df3)
```

```python
scaled_df = pd.DataFrame(scaled_data,columns=df3.columns)
```

```python
scaled_df.head()
```

```python
data = pd.concat([scaled_df,df2],axis=1)
```

```python
data.head()
```

```python
X = data.drop('RainTomorrow',axis=1)
```

```python
y = data['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(50,input_dim =25 ,activation = 'relu', kernel_initializer='he_uniform'))


model.add(Dense(50,activation = 'relu', kernel_initializer='he_uniform'))


model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model_history = model.fit(X_train,y_train,batch_size=50,validation_split=0.3,epochs=100)
```
